```python
def linear_regression(dataset,features,target):
    #division du dataset en ensemble d'entrainement et de test
    cut=round(len(dataset)*0.8)
    train_set=dataset.iloc[:cut,:].reset_index(drop=True)
    test_set=dataset.iloc[cut:,:].reset_index(drop=True)
    #entrainement du mmodel de regression linéaire
    model=linear_model.LinearRegression()
    model.fit(train_set[features],train_set[target])
    #prediction sur l'ensemble test
    predictions =model.predict(test_set[features])
    #calcule du rmse
    rmse =mean_squared_error(test_set[target],predictions)**(1/2)
    return rmse

rmse=linear_regression(datasetR,['GrLivArea'],'SalePrice')
rmse
```

Mise à jour de la fonction de prédiction pour y intégrer la standardisation des valeurs

```python
def linear_regression(datasetRoot, features, target):
    
    dataset = datasetRoot.copy()
    # Sélection des colonnes numériques à standardiser
    numerical_features=dataset.select_dtypes(include=['int64','float64']).columns.tolist()[:-1]
    
    # Standardisation des données
    scaler=preprocessing.StandardScaler()
    dataset[numerical_features]=scaler.fit_transform(dataset[numerical_features])
    
    
    cut = round(len(dataset) * 0.8)
    train_set = dataset.iloc[:cut, :].reset_index(drop=True)
    test_set = dataset.iloc[cut:, :].reset_index(drop=True)
    
    
    model = linear_model.LinearRegression()
    model.fit(train_set[features], train_set[target])
    
    
    predictions = model.predict(test_set[features])
    
    rmse =mean_squared_error(test_set[target],predictions)**(1/2)
    
    
    return rmse

# Utilisation de la fonction
```


Pourquoi Écarter Certaines Variables Catégorielles dans la Construction de notre Modèle Prédictif ?

1) une même valeur rencontrée trop souvent

- Faible Variabilité : Les variables où une valeur prédomine largement apportent peu d'informations utiles pour la prédiction. Elles ne contribuent pas de manière significative à différencier les classes ou à expliquer la variabilité de la variable cible.
- Risque de  Overfitting : Ces variables peuvent conduire le modèle à apprendre des motifs non représentatifs des données générales. Le modèle peut surapprendre des détails spécifiques au dataset d'entraînement, ce qui réduit sa capacité à généraliser sur de nouvelles données.

2) Un nombre excessif de catégories

- Complexité du Modèle : Un grand nombre de catégories peut augmenter considérablement la complexité du modèle (dummy variables). Cela peut conduire à un modèle plus lent à entraîner et plus difficile à interpréter.

Parce qu'une même valeur est rencontrée dans plus de 80 % des cas ou qu'un nombre excessif de catégories est constaté, nous supprimons les colonnes suivantes : Alley, BldgType, BsmtCond, BsmtFinType2, CentralAir, Condition1, Condition2, Electrical, ExterCond, Exterior1st, Exterior2nd, Fence, Functional, GarageCond, GarageQual, Heating, LandContour, LandSlope, MSSubClass, MiscFeature, MoSold, Neighborhood, PavedDrive, PoolQC, RoofMatl, SaleCondition, SaleType, Street.

```python
#Mise à jour de notre fonction

def linear_regression(datasetRoot, features, target):
    
    dataset = datasetRoot.copy()
    
    target_column = dataset[target]
    dataset = dataset[features].copy()
    
    numerical_features = dataset[features].select_dtypes(include=['int64', 'float64']).columns.tolist()
    scaler = preprocessing.StandardScaler()
    dataset[numerical_features] = scaler.fit_transform(dataset[numerical_features])
    
    ########### Transformation des colonnes catégorielles en variables
    categorical_features=dataset[features].select_dtypes(include=['object']).columns.tolist()
    dataset=pd.get_dummies(dataset,columns=categorical_features,drop_first=True)
    
    
    train_dataset_columns=dataset.columns
    dataset[target]=target_column
    
    ####################
    cut = round(len(dataset) * 0.8)
    train_set = dataset.iloc[:cut, :].reset_index(drop=True)
    test_set = dataset.iloc[cut:, :].reset_index(drop=True)
    
    model = linear_model.LinearRegression()
    model.fit(train_set[train_dataset_columns], train_set[target])
    predictions = model.predict(test_set[train_dataset_columns])
    rmse = mean_squared_error(test_set[target], predictions)**1/2
    
    return rmse
```

```python
# Boucle pour tester la régression linéaire avec un nombre croissant de features

fullcolumnslist = fullListNumbers + allcatColumns

newcolumns = []
result = []

for i in range(len(fullcolumnslist)):
    fit = fullcolumnslist[0:i+1]
    test = linear_regression(datasetR, fit, 'SalePrice')  # Supposons que cette fonction retourne un RMSE ou une autre métrique
    newcolumns.append(fit)
    result.append(test)

# Conversion des listes en DataFrame
df_results = pd.DataFrame({'Features': newcolumns, 'RMSE': result})

# Affichage des résultats
df_results
```
